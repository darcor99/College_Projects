```python
#Task 2
#Write a NaiveBayes function which takes the sample training and testing documents shown in the table as input, and uses Naive Bayes to classify the test document

#concat all docs to mega doc based on their class
#create bag of words: words and their probabilities of occuring within a class
#combine the 2 bags of words to get vocabulary - need size of vocabulary to calc the probabilities of a word occuring in a class

def NaiveBayes(trainingset,testset):

  testingSetLower = [(doc[0].lower(), doc[1].lower()) for doc in testset] ########## <- From Conor Pesch


  #calc priors and create bag of words
  IEcount = 0
  GBcount = 0

  WORDS_classIE = []
  WORDS_classGB = []

  for i in range(len(trainingset)):
    if trainingSet[i][1] == "GB":
      WORDS_classGB.append((trainingSet[i][0].lower()))
      GBcount +=1

    if trainingSet[i][1] == "IE":
      WORDS_classIE.append((trainingSet[i][0].lower()))
      IEcount +=1

  probIE = IEcount/len(trainingset) #initialise probabilities as prior probabilities
  probGB = GBcount/len(trainingset)
 
  print(f'probIE = {probIE}, probGB = {probGB}\n') ################New line from Conor Pesch

  WORDS_classGB = (" ".join(WORDS_classGB)).split(" ")
  WORDS_classIE = (" ".join(WORDS_classIE)).split(" ")#bag of words as list

  #make bag of words as dict with counts
  BOW_IE = {}
  BOW_GB = {}

  for word in WORDS_classIE:
    if word not in BOW_IE:
      BOW_IE[word] = 1
    else:
      BOW_IE[word] += 1

  for word in WORDS_classGB:
    if word not in BOW_GB:
      BOW_GB[word] = 1
    else:
      BOW_GB[word] += 1
          

  print(BOW_IE)
  print(BOW_GB)

  def Merge(dict1, dict2): 
    res = {**dict1, **dict2}
    return res

  VOCAB = Merge(BOW_IE,BOW_GB) #vocab
  V = len(VOCAB) #size of vocab

  print(f'V = {VOCAB}\n|V| = {V}\n')

  ############### new code below #################
  
   # iterate through docs in testing set
  for i, doc in enumerate(testingSetLower):
    print("--------------------------------------------------")
    print(f"Test Document = {testset[i]}\n")

    # find conditional probabilities for each word in testing doc
    wordConditionalProbGB = {}
    wordConditionalProbIE = {}

    for word in doc[0].split():
      if word in BOW_GB.keys():
        wordConditionalProbGB[word] = (BOW_GB[word]+1)/(len(BOW_GB)+V)
      else:
        wordConditionalProbGB[word] = (1)/(len(BOW_GB)+V)

      if word in BOW_IE.keys():
        wordConditionalProbIE[word] = (BOW_IE[word]+1)/(len(BOW_IE)+V)
      else:
        wordConditionalProbIE[word] = (1)/(len(BOW_IE)+V)

      print(f'word = "{word:15}",     wordConditionalProbGB = {wordConditionalProbGB[word]:.15},     wordConditionalProbIE = {wordConditionalProbIE[word]:.15}')

    # find class probabilities
    docProbGB = probGB * sum(list(wordConditionalProbGB.values())) #update probabilities with conditional probabilities
    docProbIE = probIE * sum(list(wordConditionalProbIE.values())) #

    print(f'\ndocProbGB = {docProbGB}, docProbIE = {docProbIE}')

    # determine class
    if docProbGB > docProbIE:
      print('Inferred class = GB')
    elif docProbIE > docProbGB:
      print('Inferred class = IE')
    else:
      print('Inferred class = IE,GB')


#test code

trainingSet = [('London is the Capital of GB','GB'),('Oxford is a city in GB','GB'),('Dublin is the capital of Ireland','IE'),('Limerick is a city in Ireland','IE')]
testSet = [('University of Limerick','?'),('University College Dublin','?'),("Imperial College London",'?'),("University of Oxford",'?'),("Ireland & GB",'?')]

NaiveBayes(trainingSet,testSet)
```
