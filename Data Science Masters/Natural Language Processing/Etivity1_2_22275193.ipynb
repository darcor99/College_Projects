```python
#2.1
!wget https://www.gutenberg.org/files/100/100-0.txt #download the completed works of Shakespeare
filePath = "/content/100-0.txt"

ShakespeareFile = open(filePath, 'r') #open the file
ShakespeareContent = ShakespeareFile.read() #read the file
ShakespeareContent = ShakespeareContent.splitlines() #split the file by each line

print(ShakespeareContent[0:50])

```

```python
#2.2

#import relevant libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import text_to_word_sequence, Tokenizer

token = Tokenizer() #make instance of Tokenizer class
token.fit_on_texts(ShakespeareContent)


Types = token.word_counts #finds all the Types present in the document 
#and creates dictionary, with Types as keys and their corresponding frequency as the values


print("There are", len(Types), "Types in Shakespeare's works") #total number of types
print(" ")

t = Types
t_sorted = {}#sort dictionary in terms of the frequency of the types
for w in sorted(t, key=t.get, reverse=True):
    t_sorted[w] = t[w]
print(t_sorted)

print(" ")


#print out most common type and its frequency
print("'",list(t_sorted.keys())[list(t_sorted.values()).index(max(t_sorted.values()))],"'"," is the most common word in this corpus and it has a frequency of ",max(t_sorted.values()), "times in Shakespeare's works")

print(" ")

values = t_sorted.values()
tokens = sum(values)#the sum of the frequency of the tokens is the no. of tokens in the Corpus 
print("The number of Tokens in Shakespeare's works is:", tokens)#prints number of tokens


```

```python
from re import S
#2.3 
#import libraries for stemming
from nltk.stem import *
from nltk.stem.porter import *
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer

print(" ")

## Stemming:
stemmer = PorterStemmer() #create instance of PorterStemmer to stem words in the document

###################Credit to Aaron Naylor for reccommending this way of finding the stemmed types to me

StemmedTypes = {} #create empty dictionary  to hold the stemmed types

#checks for all the types in the document and stems them and then saves them in the dictionary StemmedTypes
for key in Types:
  StemmedTypes[stemmer.stem(key)] = Types[key] 

####################

print("There are", len(StemmedTypes), "stemmed types")#no of stemmed types
print(" ")


##Lematizing:

lemmatizer = WordNetLemmatizer() #create instance of lemmatizer class

###################Credit to Aaron Naylor for reccommending this way of finding the lemmatized types to me

LemmatizedTypes = {} #create empty dictionary  to hold the lemmatized types

#checks for all the types in the document and lemmatizes them and then saves them in the dictionary LemmatizedTypes
for key in Types:
  LemmatizedTypes[lemmatizer.lemmatize(key)] = Types[key] 

####################

print("There are", len(LemmatizedTypes), "lemmatized types")#no of lemmatized types
print(" ")


assert len(Types) > len(LemmatizedTypes) > len(StemmedTypes),"The following statement is False: total number of types > total number of lemmatized types > total number of stemmed types"
#the following statement only gets printed if the assert statement is True
print("The following statement is True: total number of types > total number of lemmatized types > total number of stemmed types")
```
