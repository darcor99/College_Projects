## Using the Raw bigram counts from the table, we can compute the probabilities of sentences a and b. 

a = "i want to want".

b = "i want to spend".

We compute the probability of a word $w_{i}$ following a word $w_{i-1}$ using the Bigram model:
$$ P(w_{i}|w_{i-1}) = \frac{c(w_{i-1},w_{i})}{c(w_{i-1})} $$

This method finds the probability of a word $w_{i}$ following a word $w_{i-1}$ in a sentence by counting the number of times $w_{i}$ is followed by $w_{i-1}$ in the corpus divided by the count of the number of times the word $w_{i-1}$ is found in the corpus. 

A sentence can be thought of as multple bigrams. We can find the product of each bigram being present in the sentence to find the probability that the sentence occurs.

First I will calculate the probability of the sentences "a" and "b" without using Add-one (Laplace) smoothing:

Probability of Sentence a:

$$P(a) = P(\textrm{"i want to want"})$$

$$ = P(\textrm{want|i})\cdot P(\textrm{to|want}) \cdot P(\textrm{want|to})   $$

$$= \frac{\textrm{c(I, want)}}{\textrm{c(I)}} \cdot \frac{\textrm{c(want, to)}}{\textrm{c(want)}} \cdot \frac{\textrm{c(to, want)}}{\textrm{c(to)}}$$

$$= \left(\frac{827}{2533}\right)\left(\frac{608}{927}\right)\left(\frac{0}{2417}\right)$$

$$ = 0$$

Probability of Sentence b:

$$P(b) = P(\textrm{"i want to spend"})$$

$$ =   P(\textrm{want|i})\cdot P(\textrm{to|want}) \cdot P(\textrm{spend|to})   $$

$$= \left(\frac{827}{2533}\right)\left(\frac{608}{927}\right)\left(\frac{211}{2417}\right)$$

$$ = 0.01869 \approx 1.87\% $$



We alter our formula for calculating the probabilities slightly when calculating the probabilities with Add-one (Laplace Smoothing):

Now we calculate the probability of bigrams using:

$$ P'(w_{i}|w_{i-1}) = \frac{c(w_{i-1},w_{i}) + 1}{c(w_{i-1}) + V} $$
Where V is the number of Types in the Corpus (in this case it is 8)

So the probability of sentence "a" with laplace smoothing is:

$$P'(a) = P'(\textrm{"i want to want"})$$

$$ = P'(\textrm{want|i})\cdot P'(\textrm{to|want}) \cdot P'(\textrm{want|to})   $$

$$= \frac{\textrm{c(I, want)}+1}{\textrm{c(I)}+V} \cdot \frac{\textrm{c(want, to)}+1}{\textrm{c(want)}+V} \cdot \frac{\textrm{c(to, want)}+1}{\textrm{c(to)}+V}$$

$$= \left(\frac{828}{2541}\right)\left(\frac{609}{935}\right)\left(\frac{1}{2425}\right)$$

$$ = 0.000087522 \approx 0.00875 \%$$

And the probability of sentence "b" with laplace smoothing is:


$$P'(b) = P'(\textrm{"i want to spend"})$$

$$ =  \cdot P'(\textrm{want|i})\cdot P'(\textrm{to|want}) \cdot P'(\textrm{spend|to})   $$

$$= \left(\frac{828}{2541}\right)\left(\frac{609}{935}\right)\left(\frac{212}{2425}\right)$$

$$ = 0.01855 \approx 1.855 \% $$

```python
#Task 2 part (b)

import numpy as np

def probability(sentence):

  sentencelist = sentence.split()
  print(sentencelist)

  #a = unigrams_df.loc[unigrams_df[0] == sentencelist[0]] #first 0 doesnt change(1st entry in row)
  #print(a) #finds a in unigrams dataframe
  #print(int(a[1])) #count of word a
  #print("---------------------")
  
  totalunigramsfrequency = unigrams_df[1].sum()
  print("Total Unigram frequency = {0}".format(totalunigramsfrequency)) #total no of counts of all unigrams

  totalbigramsfrequency = bigrams_df[1].sum()
  print("Total Bigram frequency = {0}".format(totalbigramsfrequency))

  print("")

  sentencelen = len(sentencelist)
  #print("len = ",len(sentencelist)) #=3

  iter = 0
  problist = []
  
  for i in range(sentencelen-1):
  
    word_in_df = unigrams_df.loc[unigrams_df[0] == sentencelist[i]]
    unigramcount = int(word_in_df[1]) 
    print("unigram count for \"{0}\" = {1}".format(sentencelist[i], unigramcount))

  
    pair = sentencelist[i] + (" ") + sentencelist[i+1]
    pair = str(pair)
    pair_in_df = bigrams_df.loc[bigrams_df[0] == pair]
    bigramcount = int(pair_in_df[1])
    print("bigram count for \"{0}\" = {1}".format(pair, bigramcount))

    bigramProbability = bigramcount/unigramcount
    print("bigramprobability for \"{0}\" = {1}".format(pair,bigramProbability))
    problist.append(bigramProbability)

    print("--------------------------------------------------")
    print("")

  sentenceProbability = np.prod(problist)
  print("sentenceProbability for \"{0}\" = {1}".format(sentence,sentenceProbability))

  return sentenceProbability


assert(probability("i love you") > probability("i hate you"))
print("True")
```

```python
def probabilityLaplace(sentence):

  sentencelist = sentence.split()
  print(sentencelist)

  #a = unigrams_df.loc[unigrams_df[0] == sentencelist[0]] #first 0 doesnt change(1st entry in row)
  #print(a) #finds a in unigrams dataframe
  #print(int(a[1])) #count of word a
  #print("---------------------")
  
  totalunigramsfrequency = unigrams_df[1].sum()
  print("Total Unigram frequency = {0}".format(totalunigramsfrequency)) #total no of counts of all unigrams

  totalbigramsfrequency = bigrams_df[1].sum()
  print("Total Bigram frequency = {0}".format(totalbigramsfrequency))

  print("")

  sentencelen = len(sentencelist)
  

  iter = 0
  problist = []
  
  for i in range(sentencelen-1):
  
    word_in_df = unigrams_df.loc[unigrams_df[0] == sentencelist[i]]
    unigramcount = int(word_in_df[1] + totalUnigrams) #totalUnigrams = V, V added to unigram count for Laplace smoothing
    print("unigram count for \"{0}\" = {1}".format(sentencelist[i], unigramcount))

  
    pair = sentencelist[i] + (" ") + sentencelist[i+1]
    pair = str(pair)
    pair_in_df = bigrams_df.loc[bigrams_df[0] == pair]
    bigramcount = int(pair_in_df[1] +1) #add 1 to bigram count for Laplace Smoothing
    print("bigram count for \"{0}\" = {1}".format(pair, bigramcount))

    bigramProbability = bigramcount/unigramcount
    print("bigramprobability for \"{0}\" = {1}".format(pair,bigramProbability))
    problist.append(np.log(bigramProbability)) #log probabilities to address underflow

    print("--------------------------------------------------")
    print("")

  sentenceProbability = np.exp(np.sum(problist)) #exp of sum log probabilities
  print("sentenceProbabilityLaplace for \"{0}\" = {1}".format(sentence,sentenceProbability))

  return sentenceProbability

probabilityLaplace("i love you")

probabilityLaplace("i hate you")
```
